```python
def count_titles(csv_file_name):
    from collections import defaultdict

    raw_data = [row.split(',') for row in open(csv_file_name)]
    raw_data = raw_data[1:]
    names = set([row[0].split(' ')[-1] for row in raw_data])
    
    list_of_data = [[row[0].split(' ')[-1],row[1],row[2],row[3].strip('\n')] for row in raw_data]
    new_dict = defaultdict()
    for name in names:
        new_dict[name] = list([row[1:] for row in list_of_data if row[0] == name])
        

    
    return new_dict

filename = 'faculty.csv'
count_titles(filename)
```

```python
def get_dict():
    from collections import defaultdict

    raw_data = [row.split(',') for row in open('faculty.csv')]
    raw_data = raw_data[1:]
    list_of_data = [[tuple(row[0].split(' ')),row[1],row[2],row[3].strip('\n')] for row in raw_data]
    names = [row[0] for row in list_of_data]
    new_dict = defaultdict()
    for name in names:
        new_dict[name] = [row[1:] for row in list_of_data if row[0] == name][0]
         
    return dict(new_dict)

filename = 'faculty.csv'
count_titles(filename)
```
